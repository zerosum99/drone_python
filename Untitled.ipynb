```python
from time import sleep

from e_drone.drone import *
from e_drone.protocol import *

from serial.tools.list_ports import comports

for port, desc, hwid in sorted(comports()):
    print("%s" % (port))


if __name__ == '__main__':

    drone = Drone()
    drone.open(port)

    print("TakeOff")
    drone.sendTakeOff()
    for i in range(5, 0, -1):
        print("{0}".format(i))
        sleep(1)

    print("Hovering")
    pitch = 1
    for i in range(4, 0, -1):
        print("{0}".format(i))
        pitch +=  i
            
        drone.sendControlWhile(pitch, 0, 0, 0, 1000)
        sleep(0.01)

    print("Landing")
    drone.sendLanding()
    for i in range(5, 0, -1):
        print("{0}".format(i))
        sleep(1)

    drone.close()
```
